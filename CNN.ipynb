```python
import torch
import numpy as np
```

```python
def conv1D(x, w, p = 0, s = 1):
    w_rot = np.array(w[::-1])
    x_padded = np.array(x)
    if p > 0:
        padded_arr = np.zeros(p)
        x_padded = np.concatenate([padded_arr , x_padded , padded_arr])
    res = []
    for i in range(0, int(len(x_padded) - len(w_rot)) + 1, s):
        res.append(np.sum(x_padded[i:i + w_rot.shape[0]] * w_rot))
    return np.array(res)
```

```python
x = [1, 3, 2, 4, 5, 6, 1, 3]
```

```python
w = [1, 0, 3, 1, 2]
```

```python
print('Conv1d Implementation:',
... conv1D(x, w, p=2, s=1))
```

```python
print('NumPy Results:',
... np.convolve(x, w, mode='same'))
```

```python
import scipy.signal 
```

```python
def conv2D(X,W, p =(0,0), s=(1,1)):
    W_rot = np.array(W)[::-1,::-1]
    X_org = np.array(X)
    n1 = X_org.shape[0]+ 2*p[0]
    n2 = X_org.shape[1]+ 2*p[1]
    X_padded = np.zeros((n1,n2))
    X_padded[p[0]:p[0]+X_org.shape[0], p[1] : p[1] +X_org.shape[1]] = X_org
    res = []
    for i in range(0, int((X_padded.shape[0] - W_rot.shape[0])/s[0]) + 1, s[0]):
        res.append([])
        for j in range(0, int((X_padded.shape[1] - W_rot.shape[1]) / s[1]) + 1, s[1]):
            X_sub = X_padded[i:i + W_rot.shape[0], j:j + W_rot.shape[1]]
            res[-1].append(np.sum(X_sub*W_rot))
    return np.array(res)
```

```python
X = [[1, 3, 2, 4], [5, 6, 1, 3], [1, 2, 0, 2], [3, 4, 3, 2]]
```

```python
W = [[1, 0, 3], [1, 2, 1], [0, 1, 1]]
```

```python
print('Conv2d Implementation:\n',conv2D(X, W, p=(1, 1), s=(1, 1)))
```

```python
print('SciPy Results:\n',scipy.signal.convolve2d(X, W, mode='same'))
```
