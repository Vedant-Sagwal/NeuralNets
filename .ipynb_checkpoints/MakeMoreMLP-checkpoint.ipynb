```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
chars = sorted(list(set("".join(words))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi["."] = 0
itos = {i:s for s,i in stoi.items()}
print(stoi)
print("----------------------")
print(itos)
```

```python
#build dataset
block_size = 3
X = []
Y = []
context = [0] * block_size
for word in words:
    # print(word)
    for ch in word+".":
        ix = stoi[ch]
        X.append(context)
        Y.append(ix)
        context = context[1:] + [ix]
        # print("".join([itos[i] for i in context]), "------>", itos[ix])
X = torch.tensor(X)
Y = torch.tensor(Y)
```

```python
#With generator
n = X.shape[0]
g = torch.Generator().manual_seed(2147483647)
C = torch.randn((27, 10),generator = g)
W1 = torch.randn((30, 200),generator = g)
b1 = torch.randn(200,generator = g)
W2 = torch.randn((200, 27),generator = g)
b2 = torch.randn(27, generator = g)
parameters = [C, W1,W2,b1,b2]
for p in parameters :
    p.requires_grad = True
```

```python
lre = torch.linspace(-3, 0, 1000)
lrs = 10**lre
```

```python
# lri = []
# lossi = []
for i in range(100000):
    ix = torch.randint(0,X.shape[0], (32,)) # create a mini batch of size 32 of randomly selected int from X on which grad wioll be calculateed
    #forward passs
    emb = C[X[ix]] #[batchsize,3,2]
    h = torch.tanh(emb.view((-1, 6)) @ W1 + b1)
    logits = h @ W2 + b2
    loss = F.cross_entropy(logits,Y[ix])
    #zero_grad
    for p in parameters:
        p.grad = None
    #backward prop
    loss.backward()
    #updation of grad
    # lr = lrs[i]
    # from below graph we can see 10 ** -1 is a pretty good learning rate so lr =0.1
    lr = 0.1
    for p in parameters:
        p.data += -lr * p.grad
    # lossi.append(loss.item())
    # lri.append(lre[i])
print(loss.item())
```

```python
#loss on full dataset
emb = C[X]
h = torch.tanh(emb.view((-1, 6)) @ W1 + b1)
logits = h @ W2 + b2
loss = F.cross_entropy(logits,Y)
loss
```

# Splitting the dataset to training, testing and validation dataset

```python
import random 
# 80 train, 10 validation, 10 test
#build the dataset
# chars = sorted(list(set("".join(words))))
# stoi = {s:i+1 for i,s in enumerate(chars)}
# stoi["."] = 0
# itos = {i:s for s,i in stoi.items()}
def build_dataset(words):
    X = []
    Y = []
    block_size = 3
    for word in words:
        context = [0] * block_size
        for ch in word+".":
            ix = stoi[ch]
            X.append(context)
            Y.append(ix)
            context = context[1:] + [ix]
    X = torch.tensor(X)
    Y = torch.tensor(Y)
    print(X.shape, Y.shape)
    return X,Y
random.seed(42)
random.shuffle(words)
n1 = int(0.8* len(words))
n2 = int(0.9 * len(words))

Xtr, Ytr = build_dataset(words[:n1])
Xdev, Ydev = build_dataset(words[n1:n2])
Xtst, Ytst = build_dataset(words[n2:])
```

```python
for i in range(50000):
    ix = torch.randint(0,Xtr.shape[0], (32,)) # create a mini batch of size 32 of randomly selected int from X on which grad wioll be calculateed
    #forward passs
    emb = C[Xtr[ix]] #[batchsize,3,2]
    h = torch.tanh(emb.view((-1, 30)) @ W1 + b1)
    logits = h @ W2 + b2
    loss = F.cross_entropy(logits,Ytr[ix])
    #zero_grad
    for p in parameters:
        p.grad = None
    #backward prop
    loss.backward()
    #updation of grad
    # lr = lrs[i]
    # from below graph we can see 10 ** -1 is a pretty good learning rate so lr =0.1
    lr = 0.01
    for p in parameters:
        p.data += -lr * p.grad
    # lossi.append(loss.item())
    # lri.append(lre[i])
# print(loss.item())
```

```python
emb = C[Xtr]
h = torch.tanh(emb.view((-1, 30)) @ W1 + b1)
logits = h @ W2 + b2
loss = F.cross_entropy(logits,Ytr)
print(loss.item())
```

```python
emb = C[Xdev]
h = torch.tanh(emb.view((-1, 30)) @ W1 + b1)
logits = h @ W2 + b2
loss = F.cross_entropy(logits,Ydev)
print(loss.item())
```

```python
for _ in range(20):
    context = [0] * block_size
    out = []
    while True:
        emb = C[torch.tensor([context])]
        h = torch.tanh(emb.view(1, -1) @ W1 + b1)
        logits = h@W2 + b2
        probs = F.softmax(logits,dim=1)
        ix = torch.multinomial(probs, num_samples=1, generator = g).item()
        context = context[1:] + [ix]
        out.append(ix)
        if ix == 0:
            break
    print("".join(itos[i] for i in out))
```
